## Pandas DataFrames

Pandas DataFrames make manipulating your data easy, from selecting or replacing columns and indices to reshaping your data. It also allows many types (unlike numpy arrays)

Since dataframes are in the pandas library we need to import the pandas library:

```python
import pandas as pd
```

### CSV file -> Data Frame

Let's set up the data that we will be using before we start exploring Python. Here, we're going to import the data from `Classroom-Large-Sample.csv` as a Pandas dataframe called `df`.

```python
#load (read) the csv file and save into a variable called df
df = pd.read_csv("Classroom-Large-Sample.csv")
```

Unfortunately pandas can't work with column names that have spaces. This is a very common problem when importing a csv file. The way we deal with this is by replacing the space with an underscore `_`:

```python
# replace the space with an underscore _
df.columns = df.columns.str.strip().str.replace(' ', '_')

#show updated df
df
```

## Plotting
In this section, we will explore various type of plots in Python. To do this, we need to first call the matplotlib plotting library:

```python
import matplotlib.pyplot as plt
```

### Histogram

```python
plt.hist(df.Age, bins = 3)
plt.show()
```

```python
plt.hist(df.Age, bins = 5)
plt.show()
```

### Line Plots

```python
#Group by Gender
groupby_gender = df.groupby(by=['Gender'])
```

```python
#Mean scores by gender
gender_data_avg = groupby_gender.mean()
gender_data_avg
```

```python
gender_data_avg.plot()
```

### Scatter Plots

```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns

#Let's see how the Test Preparation Course affects the marks
sns.lmplot("Science_Score", "Math_Score", data=df, hue="Test_Preparation_Course", fit_reg=False, col='Gender', col_wrap=2)
```

```python
sns.lmplot("Science_Score", "Math_Score", data=df, hue="Parental_Level_of_Education", fit_reg=False, col='Gender', col_wrap=2)
```

```python
sns.lmplot("English_Score", "Math_Score", data=df, hue="Parental_Level_of_Education", fit_reg=False, col='Gender', col_wrap=2)
```

```python
#Let's see how the Test Preparation Course affects the marks
sns.lmplot("Science_Score", "Math_Score", data=df, hue="Test_Preparation_Course", fit_reg=False, col='Gender', col_wrap=2)
```

### Seaborn Violin Plots
These plots are excellent for dealing with large continuous datasets, and can similarly be segmented by an index. Using our student dataset we can gain a greater understanding about the score distribution. 

We want to know how the score (y axis) is affected by Test Preparation Course and Parents' level of education (x axis). 

 We can then set our hue to separate out the students' gender

```python
sns.violinplot('Test_Preparation_Course', "Science_Score", data=df, hue='Gender')
```

```python
sns.violinplot('Parental_Level_of_Education', "Science_Score", data=df, hue='Gender')
```

```python
sns.violinplot('Test_Preparation_Course', "English_Score", data=df, hue='Gender')
```

```python
sns.violinplot('Parental_Level_of_Education', "English_Score", data=df, hue='Gender')
```

```python
sns.violinplot('Test_Preparation_Course', "Math_Score", data=df, hue='Gender')
```

```python
sns.violinplot('Parental_Level_of_Education', "Math_Score", data=df, hue='Gender')
```

### Pairplots & Correlation Heatmaps
In datasets with a small number of features or columns (less than 15) Seaborn Pairplots can quickly enable a visual inspection of any relationships between variables. Graphs along the left diagonal represent the distribution of each feature, whilst graphs on off diagonals show the relationship between variables.

```python
sns.pairplot(df)
```

```python
#Plot based on Test Preparation Course
sns.pairplot(df, hue = 'Test_Preparation_Course')
```

```python
#Plot based on Test Preparation Course
sns.pairplot(df, hue = 'Gender')
```

```python
#Plot based on Test Preparation Course
sns.pairplot(df, hue = 'Parental_Level_of_Education')
```
